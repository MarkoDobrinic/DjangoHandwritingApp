```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
res = np.load('result.npy')
res = res.astype(np.float32)
res[:,1] = -res[:,1]
```

```python
plt.figure(figsize=(20, 4))
plt.scatter(res[:, 0], res[:, 1])
```

```python
x_mean = np.mean(res[:, 0])
y_mean = np.mean(res[:, 1])

x_std = np.std(res[:, 0])
y_std = np.std(res[:, 1])

#res[:,0 ] = (res[:,0]-1)
#res[:,1] = (res[:,1]-1)
res[:,0:2] = res[:,0:2]/np.std(res[:,0:2], axis=0)
```

```python
plt.figure(figsize=(20, 4))
plt.scatter(res[:, 0], res[:, 1])
```
